```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
car = pd.read_csv("car.csv", encoding = "CP949")
```

```python
car
```

```python
cargroup = car.groupby("년도")[['승용', '승합', '화물', '특수']].sum()
```

```python
print(cargroup)
```

```python
year = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
```

```python
fig, ax = plt.subplots(figsize=(20,6))
bar_width = 0.2

index = np.arange(9)

b1 = plt.bar(index, cargroup['승용'], bar_width, alpha=0.4, color='red', label='Cars')
b2 = plt.bar(index + bar_width, cargroup['승합'], bar_width, alpha=0.4, color='blue', label='Vans')
b3 = plt.bar(index + 2 * bar_width, cargroup['화물'], bar_width, alpha=0.4, color='green', label='Trucks')
b4 = plt.bar(index + 3 * bar_width, cargroup['특수'], bar_width, alpha=0.4, color='orange', label='Specials')

plt.xticks(np.arange(bar_width, 9 + bar_width, 1), year)

plt.xlabel('years', size = 13)
plt.ylabel('count', size = 13)
plt.legend(fontsize = 11)
plt.show()
```
